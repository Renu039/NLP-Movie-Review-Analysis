```python
def get_cast(cast_dict):
    temp = []
    cast_dict = eval(cast_dict)
    counter = 1
    for cast in cast_dict:
        if counter <= 4:
            temp.append(cast['name'].replace(' ',''))
        else:
            break
        counter+=1
    return temp
    
```

```python
movies['cast'] = movies['cast'].apply(get_cast)
```

```python
movies.head()
```

```python
def get_crew(crew_dict):
    temp=set()
    crew_dict = eval(crew_dict)
    for crew in crew_dict:
        if crew['job']=='Director' or crew['job']=='Writer':
            temp.add(crew['name'].replace(' ',''))
    return list(temp)        
```

```python
movies['crew'] = movies['crew'].apply(get_crew)
```

```python
def get_genres(genres_dict):
    temp=[]
    genres_dict = eval(genres_dict)
    for genres in genres_dict:
        # if genres['name']in ['Action','Adventure','Fantasy','Science Fiction','Crime','Thriller','Animation','Family']:
        temp.append(genres['name'].replace(' ',''))
    return temp
```

```python
movies['genres'] = movies['genres'].apply(get_genres)
```

```python
def get_keywords(keywords_dict):
    temp=[]
    keywords_dict = eval(keywords_dict)
    for keywords in keywords_dict:
        temp.append(keywords['name'].replace(' ',''))
    return temp
```

```python
import string
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
```

```python
def remove_stopwords_simple(text):
    language='english'
    stop_words = set(stopwords.words(language))
    return ' '.join([word for word in str(text).split() if word.lower() not in stop_words])
```

```python
def remove_punctuation(text):
    temp = []
    punctuations = list(string.punctuation)
    for word in text:
        for symbol in punctuations:
            if symbol in word:
                word = word.replace(symbol,'')
        temp.append(word)
    return ' ' .join(temp)
```

```python
movies['Tags'] = movies['Tags'].apply(remove_punctuation)
```

```python
# remove_stopwords_simple(movies.Tags[0])
```

```python
movies['Tags'] = movies['Tags'].apply(remove_stopwords_simple)
```

```python
movies['Tags'] = movies['Tags'].apply(remove_stopwords_simple)
```

```python
from nltk.stem.porter import PorterStemmer

porter_stemmer = PorterStemmer()
```

```python
def stem_words(text):
    text = text.split()
    temp = []
    for word in text:
        word = porter_stemmer.stem(word)
        temp.append(word)
    return ' '.join(temp)
```

```python
movies['Tags'] = movies['Tags'].apply(stem_words)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(max_features=1000,stop_words='english')
```

```python
tfidf.fit(movies['Tags'])
```

```python
movie_vector = tfidf.transform(movies['Tags']).toarray()
```

```python
movie_vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similar_movies = cosine_similarity(movie_vector)
```

```python
def get_recommendation(movie_name):
    try:
        movie_index = movies[movies['original_title']==movie_name].index[0]
        movie_vector = similar_movies[movie_index]
        movie_vector = list(enumerate(movie_vector))
        sorted_movie_vector = sorted(movie_vector,reverse=True,key=lambda value:value[1])
        five_similar_movie_vector = sorted_movie_vector[1:6]
        for index,similarity in five_similar_movie_vector:
            movie_name = movies.iloc[index]['original_title']
            similarity = round(similarity*100,2)
            print(f" '{movie_name}': {similarity}% Similar to your recently watched movie.")
    except:
        print('Please give a proper movie title.')
  
```
